```python
# import Dependencies
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
import time
from pprint import pprint as p

# Scrape NASA Mars news site
url = "https://mars.nasa.gov/news/"
data = requests.get(url).text
soup = bs(data, 'lxml')

# print soup to see scraped data
print(soup.prettify())
```

```python
# Set first news article and paragraph sentence from soup
news_title = soup.find(class_="content_title").text.strip()
news_p = soup.find(class_="rollover_description_inner").text.strip()
print(news_title)
print(news_p)
```

```python
# Scrape NASA Mars news site for featured image
url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
img_url = "https://www.jpl.nasa.gov/"
data = requests.get(url).text
soup = bs(data, 'html.parser')

# print soup to see scraped data
print(soup.prettify())
```

```python
# image is in footer
footer = soup.find(class_="button fancybox")
img_link = footer["data-fancybox-href"]

# Combine image url and image link
featured_image_url = img_url + img_link

print(featured_image_url)
```

```python
# Scrape Mars Twitter for weather
url = "https://twitter.com/marswxreport?lang=en"
data = requests.get(url).text
soup = bs(data, 'html.parser')
print(soup.prettify())
```

```python
# set text of latest weather tweet to a variable
mars_weather = soup.find(class_="TweetTextSize").text
print(mars_weather)
```

```python
# Scrape website for Mars facts using pandas
# use pandas to convert
url = "https://space-facts.com/mars/"
tables = pd.read_html(url)
facts = tables[0]
facts = facts.rename(columns={0:"Metric", 1:"Value"})
facts_html = facts.to_html(index=False)
p(facts_html)
```

```python
# Scrape photos of Mars hemisphere

# set up chrome path for splinter
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

# make initial call to Website to get titles
data = requests.get(url).text
soup = bs(data, 'html.parser')
titles = soup.find_all("h3")

# start chrome browser
browser.visit(url)
time.sleep(3)

# list to hold dictionaries
hemisphere_image_urls = []

for i in titles:
    browser.click_link_by_partial_text(i.text[:12])
    time.sleep(2)
    html = browser.html
    soup = bs(html, 'html.parser')
    li = soup.find(class_="downloads").find("li").a["href"]
    tempDict = {}
    tempDict["title"] = i.text
    tempDict["img_url"] = li
    hemisphere_image_urls.append(dict(tempDict))
    browser.back()
    time.sleep(2)

for i in hemisphere_image_urls:
    print(i)
```
